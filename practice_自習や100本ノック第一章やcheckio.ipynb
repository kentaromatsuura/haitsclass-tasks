```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
from sklearn.datasets import load_boston # bostonデータを格納したクラス
boston = load_boston()                   # インスタンスを生成
columns = boston.feature_names
columns
df_data = pd.DataFrame(boston.data, columns=boston.feature_names)
df_data.head()
df_target = pd.DataFrame(boston.target, columns=['MEDV'])
df_target.head()
df=pd.concat([df_data,df_target],axis=1)
df.head()
df_pickup= df.loc[:, ['LSTAT', 'INDUS', 'DIS', 'RM', 'MEDV']]
import seaborn as sns
df.corr()
plt.figure(figsize=(12,9))
sns.heatmap(df.corr(),annot=True,square=True,fmt='.2f')
plt.show()
```

```python
X = df.loc[:, ['LSTAT', 'RM']].values
X
y = df.loc[:, ['MEDV']].values

y[:10]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.intercept_
# 偏回帰係数を出力
# [LSTATの係数, RMの係数]
lr.coef_
```

```python
print('R^2')
print('train: %.3f' % lr.score(X_train, y_train))
print('test : %.3f' % lr.score(X_test, y_test))
```

```python
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
# 自由度調整済み決定係数を出力
print('adjusted R^2')
print('train: %3f' % adjusted(lr.score(X_train, y_train), len(y_train), 2))
print('test : %3f' % adjusted(lr.score(X_test, y_test), len(y_test), 2))
```

```python
class Character:
    def __init__(self,name,maxhp,attack_point):
        self.name = name
        self.maxhp = maxhp 
        self.hp = maxhp
        self.attack_point = attack_point
        
    def status(self):
        return "%s:体力 %d/%d:攻撃力 %d"%(self.name,self.hp,self.maxhp,self.attack_point)
    
    def attack(self,enemy):
        enemy.hp -= self.attack_point
        print("*%sの攻撃！%sに%dのダメージ！"%(self.name,enemy.name,self.attack_point))

# インスタンスを初期化
yusha = Character("勇者",60,10)
slime = Character("スライム",15,5)

# ステータスを表示
print("------------------------------------")
print(yusha.status())  # 勇者のステータス
print(slime.status())  # スライムのステータス
print("------------------------------------")
    
# 勇者の攻撃
yusha.attack(slime)
# スライムの攻撃:
slime.attack(yusha)

# 空行を表示
print("")
print("")

# ステータスを表示
print("------------------------------------")
print(yusha.status())  # 勇者のステータス
print(slime.status())  # スライムのステータス
print("------------------------------------")
```
