```python
X = df_train.loc[:, ['OverallQual', 'GrLivArea']].values
X
```

```python
y= df_train.loc[:, ['SalePrice']].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print('R^2')
print('train: %.3f' % lr.score(X_train, y_train))
print('test : %.3f' % lr.score(X_test, y_test))
```

```python
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
print('adjusted R^2')
print('train: %3f' % adjusted(lr.score(X_train, y_train), len(y_train), 2))
print('test : %3f' % adjusted(lr.score(X_test, y_test), len(y_test), 2))
#linearだとまだ少し決定係数が低いので、多項式回帰を行う
```

```python
from sklearn.metrics import mean_squared_error as mse
print('RMSE')
print('train: %.3f' % (mse(y_train, lr.predict(X_train)) ** (1/2)))
print('test : %.3f' % (mse(y_test, lr.predict(X_test)) ** (1/2)))
#重回帰分析だとかなり
```

```python
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)    # 2次の多項式規定を生成
X_quad = quad.fit_transform(X) # 生成した基底関数で変数変換を実行
X_quad
```

```python
X_quad_train, X_quad_test, y_train, y_test = train_test_split(X_quad, y, test_size = 0.3, random_state = 0)
model_quad = LinearRegression()

# 変換後の変数を利用して実行
model_quad.fit(X_quad_train, y_train)
```

```python
cubic = PolynomialFeatures(degree=3)     # 3次の多項式規定を生成
X_cubic = cubic.fit_transform(X) # 生成した基底関数で変数変換を実行
X_cubic_train, X_cubic_test, y_train, y_test = train_test_split(X_cubic, y, test_size = 0.3, random_state = 0)
# 線形回帰のインスタンスを生成
model_cubic = LinearRegression()

# 変換後の変数を利用して実行
model_cubic.fit(X_cubic_train, y_train)
```

```python
print('model_quad')
print('train: %.3f' % adjusted(model_quad.score(X_quad_train, y_train), len(y_train), 2))
print('test : %.3f' % adjusted(model_quad.score(X_quad_test, y_test), len(y_test), 2))
print('')

print("model_cubic")
print('train: %.3f' % adjusted(model_cubic.score(X_cubic_train, y_train),len(y_train),3))
print('test: %.3f' %adjusted(model_cubic.score(X_cubic_test, y_test),len(y_test),3))

#テストデータの決定係数が0.7を超えたので重要な特徴量は'OverallQual', 'GrLivArea'で、分析手法としては複数説明変数での多項式回帰である。特に
```

```python
print('RMSE_quad')
print('train: %.3f' % (mse(y_train, model_quad.predict(X_quad_train)) ** (1/2)))
print('test : %.3f' % (mse(y_test, model_quad..predict(X_quad_test)) ** (1/2)))

print('RMSE_cubic')
print('train: %.3f' % (mse(y_train, model_cubic.predict(X_cubic_train)) ** (1/2)))
print('test : %.3f' % (mse(y_test, model_cubic.predict(X_cubic_test)) ** (1/2)))
```
